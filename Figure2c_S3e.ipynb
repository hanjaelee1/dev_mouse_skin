## Cluster composition over time

```python
adata_subset = adata[adata.obs['sub_cluster'].isin(['BK-1', 'emFIB-2', 'FIB-1', 'FIB-2', 'emFIB-3', 'FIB-3', 'emFIB-1', 'BK-2', 'FIB-4', 'emBK', 'DC/DP', 'DP', 'HFKC-1', 'HFKC-2', 'SBK-1', 'HFKC-3', 'emFIB-4', 'emSBK', 'DS-2', 'DS-1', 'APM', 'HFKC-4', 'SBK-2'])]
```

```python
tmp = pd.crosstab(adata_subset.obs['sub_cluster'],adata.obs['bulk.ident'], normalize='index').reindex(['emFIB-4', 'emFIB-3', 'emFIB-2', 'emFIB-1', 'FIB-1', 'FIB-4', 'FIB-2', 'FIB-3', 'DC/DP', 'DS-2', 'APM', 'DS-1', 'DP', 'emBK', 'emSBK', 'BK-2', 'BK-1', 'SBK-1', 'SBK-2', 'HFKC-1', 'HFKC-4', 'HFKC-3',  'HFKC-2'])
tmp.plot.bar(stacked=True).legend(loc='upper right')
```
